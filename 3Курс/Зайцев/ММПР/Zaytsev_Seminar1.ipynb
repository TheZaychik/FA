# Задача про овощи

```python
from scipy.optimize import linprog
```

```python
obj = [60, 35, 55, 90]

lhs_ineq = [[-35, -20, -3, -50],
            [-4, -4, -0.5, -30],
            [-80, -60, -40, -120]]
rhs_ineq = [-1500,
            -450 ,
            -2500]

```

```python
bnd = [(0, 30),
       (0, 30),
       (0, 30),
       (0, 30)]

opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq,
              bounds=bnd,
              method="highs") #revised simplex устаревший, программа советует этот
```

```python
opt
```

```python
opt.fun
```

```python
opt.x
```

Программа предлагает есть 30 кг картофеля за 900р

# КомпМагазин

```python
obj = [-21, -49, -164, -1353]

lhs_ineq = [[1, 1, 5, 30],
            [1, 1, 5, 30],
            [1, 3, 25, 25],
            [1, 8, 25, 50],
            [1, 2, 3, 10],
            [1, 1, 5, 12],
            [8, 12, 25, 30]]

rhs_ineq = [200,
            500,
            250,
            400,
            300,
            400,
            800]
```

```python
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq, method="highs", integrality=1)
```

```python
opt
```

```python
opt.fun
```
