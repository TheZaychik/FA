**Импорт и настрока matplotlib**

```python
import torch
import matplotlib
import matplotlib.pyplot as plt
```

```python
matplotlib.rcParams['figure.figsize'] = (13.0, 5.0)
```

**Тренировочное множество**

```python
x_train = torch.rand(100)
x_train = x_train * 20.0 - 10.0

y_train = torch.sin(x_train)

plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('$y = sin(x)$')
```

Проверим размерности train множества

```python
x_train.shape, y_train.shape
```

Создание шума для выходных значений

```python
noise = torch.randn(y_train.shape) / 5

plt.plot(x_train.numpy(), noise.numpy(), 'o')
plt.axis([-10, 10, -1, 1])
plt.title('Gaussian noise')
```

Добавляем шум к выходным значениям

```python
y_train += noise
```

```python
plt.plot(x_train.numpy(), y_train.numpy(), 'o')
plt.title('noisy sin(x)')
plt.xlabel('x_train')
plt.ylabel('y_train')
```

Проверочное множество

```python
x_val = torch.linspace(-10,10,100)
y_val = torch.sin(x_val.data)
plt.plot(x_val.numpy(), y_val.numpy(), 'o')
plt.title('val sin(x)')
plt.xlabel('x_val')
plt.ylabel('y_val')
```
