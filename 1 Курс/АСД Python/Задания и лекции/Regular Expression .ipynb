
1) Дана строка 'aaa--bbb==ccc__ddd'. Написать регулярное выражение для разбивки строки на список ['aaa','bbb','ccc','ddd'].

```python
import re
a = re.compile(r'[-_=]+')
m = re.split(a, 'aaa--bbb==ccc__ddd')
print(m)
```

2) Получить первое слово из произвольной строки. Например, для 'Yesterday, All my troubles seemed so far away' должно возвращаться 'Yesterday'.

```python
text = re.compile(r'\w+')
print(text.findall('Yesterday, All my troubles seemed so far away')[0])
```

6) Проверить корректность введенного E-mail

```python
text = 'aaa@mail.ru'
regex = re.compile(r'[\w|\.?]*[@{1}]\w*[\.]{1}\w*')
print(re.match(regex, text))
```
